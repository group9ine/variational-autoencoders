### Distance Matrix

```python
def calculate_distance_matrices(coordinates):
    num_configs = coordinates.shape[0]
    distance_matrices = []

    for i in range(num_configs):
        config = coordinates[i]
        distance_matrix = cdist(config, config, metric='euclidean')
        distance_matrices.append(distance_matrix)
    
    return distance_matrices
```

```python
dist_matrices = np.array(calculate_distance_matrices(vcs))
```
