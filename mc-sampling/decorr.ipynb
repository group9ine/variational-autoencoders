```python
import numpy as np
import matplotlib.pyplot as plt

# Generate a sample list of data (list of arrays)
data = [np.random.uniform(0, 1, 60) for i in range(100)]
wsl_path = r'\\wsl$\Ubuntu\home\alepitte\ale\uni\pos.txt'
data = np.loadtxt(wsl_path)
```

```python
data
```

```python
max_corr = float( np.correlate(data[0], data[1]) )
max_corr
```

```python
acf_list = []
c = 0
t = []
for k, config in enumerate(data):
    acf = np.correlate(data[c], config)/max_corr
    acf_list.append(acf)
    if acf<0.1:
        c=k
        t.append(c)
        
```

```python
dt = [t[i+1]-t[i] for i in range(len(t)-1)]
np.mean(dt) #mean decorr time
```

```python
fig, ax = plt.subplots(figsize=(40,7))
ax.plot(acf_list, )
```
