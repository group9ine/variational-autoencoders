```python
import numpy as np
import matplotlib.pyplot as plt

# Generate a sample list of data (list of arrays)
data = [np.random.uniform(0, 1, 60) for i in range(100)]
wsl_path = r'\\wsl$\Ubuntu\home\alepitte\ale\uni\pos.txt'
data = np.loadtxt(wsl_path)
```

```python
data
```

```python
# Calculate the autocorrelation function for each configuration
acf_list = []
for config in data[100000:]:
    acf = np.correlate(data[100000], config)
    acf_list.append(acf)
acf_list = acf_list / np.max(acf_list)

# Find the first index where the autocorrelation drops below 0.1 (i.e. 10% of its maximum value)
autocorr_times = []
for acf in acf_list:
    autocorr_time = np.where(acf < 0.1)
    autocorr_times.append(autocorr_time)


print("Autocorrelation times:", autocorr_time)
acf_list
```
